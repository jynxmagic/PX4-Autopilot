```python
HOME_DIR = "/home/lunet/ttcc4"
def load_df_from_episode(run, time_s):

    # print("-Loading csvs into pandas dataframes")
    #load csv
    files_required = [
        "vehicle_attitude_0.csv",
        "vehicle_control_mode_0.csv"
    ]

    columns = ["timestamp", "q[0]", "q[1]", "flag_control_ai_enabled", "q[2]", "q[3]"]
    prefix = "./"+time_s+"_"
    csvs = []
    for filename in files_required:
        csvs.append(pd.read_csv(prefix+filename))
    
    df = pd.concat(csvs)[columns]
    min_ai_enabled = df[df["flag_control_ai_enabled"] == 1].timestamp.min()
    df = df[df.timestamp >= min_ai_enabled]
    df["run"] = run
    df = df[df["q[0]"].notna()]

    q0 = df["q[0]"]
    q1 = df["q[1]"]
    q2 = df["q[2]"]
    q3 = df["q[3]"]

    df["roll"] = np.arctan2(2.0 * (q0 * q1 + q2 * q3),
                        1.0 - 2.0 * (q1 * q1 + q2 * q2))
    df["pitch"] = np.arcsin(2.0 * (q0 * q2 - q3 * q1))

    df["yaw"] = np.arctan2(2.0 * (q0 * q3 + q1 * q2),
                        1.0 - 2.0 * (q2 * q2 + q3 * q3))
    df["yaw"] -= 1.57


    df.drop(columns=["q[0]", "q[1]", "q[2]", "q[3]"], inplace=True)
    df.sort_values(by="timestamp", inplace=True)
    df.reset_index(inplace=True)
    return df
```

```python
HOME_DIR = "/home/lunet/ttcc4"
def load_df_from_episode(run, time_s):
    #load csv
    files_required = [
        "vehicle_local_position_0.csv",
        "vehicle_control_mode_0.csv"
    ]

    columns = ["timestamp", "vx", "vy", "flag_control_ai_enabled", "vz"]
    prefix = "./"+time_s+"_"
    csvs = []
    for filename in files_required:
        csvs.append(pd.read_csv(prefix+filename))
    
    df = pd.concat(csvs)[columns]
    min_ai_enabled = df[df["flag_control_ai_enabled"] == 1].timestamp.min()
    df = df[df.timestamp >= min_ai_enabled]
    df["run"] = run
    df = df[df["vx"].notna()]

    df.vz = -df.vz
    df["LVErr"] = df.vx ** 2 + df.vy ** 2 ** 2  + df.vz ** 2
    df.sort_values(by="timestamp", inplace=True)# rewards = pd.DataFrame({"Timestep":rewards.index,"Reward":rewards})
    df.reset_index(inplace=True)
    return df
```

```python
HOME_DIR = "/home/lunet/ttcc4"
def load_df_from_episode(run, time_s):
    #load csv
    files_required = [
        "actuator_motors_0.csv",
        "vehicle_control_mode_0.csv"
    ]

    columns = ["timestamp", "control[0]", "control[1]", "flag_control_ai_enabled", "control[2]",  "control[3]"]
    prefix = "./"+time_s+"_"
    csvs = []
    for filename in files_required:
        csvs.append(pd.read_csv(prefix+filename))
    
    df = pd.concat(csvs)[columns]
    min_ai_enabled = df[df["flag_control_ai_enabled"] == 1].timestamp.min()
    df = df[df.timestamp >= min_ai_enabled]
    df["run"] = run
    df = df[df["control[0]"].notna()]

    df["ControlErr"] = (df["control[0]"] - 0.63) ** 2 +  (df["control[1]"] - 0.63) ** 2  +  (df["control[2]"] -0.63) ** 2 +  (df["control[3]"] - 0.63)  ** 2
    df.sort_values(by="timestamp", inplace=True)
    df.reset_index(inplace=True)
    return df
```
